```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
d1 = sns.load_dataset('iris')
d1.sample(4)
```

```python
plt.scatter(d1['sepal_width'],d1['petal_length'])
plt.title("Used iris dataset",fontsize = 20, color = 'Green')
plt.grid(color = 'grey')
plt.show()
```
