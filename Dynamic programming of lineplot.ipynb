```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
n1 = np.array(list(map(int, input('Enter data 1:').split())))
n2 = np.array(list(map(int, input('Enter data 2:').split())))
plt.plot(n1,n2,marker ='*')
plt.title("Points scored by students\n",fontsize = 20, color = 'Green')
plt.grid(color = 'grey')
plt.xlabel("Number of Points")
plt.ylabel("Number of Students")
plt.show()
```
