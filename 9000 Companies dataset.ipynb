```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
l1 = pd.read_csv("company_dataset.csv")
l1.sample(10)
```

```python
l2 = l1[l1['ctype'].isin(['private', 'public'])]
```

```python
print(l1.dtypes)
print(l1.head())
```

# <span style = color:green> Line chart
***

```python
l2 = l1.copy()
l2['ctype'] = l2['ctype'].astype(str).str.lower()
l2 = l2[l2['ctype'].isin(['private', 'public'])]
```

```python
mapping = {'private': 0, 'public': 1}
l2['ctype_num'] = l2['ctype'].map(mapping)
```

```python
l2_10 = l2.head(10)
mapping = {'private': 0, 'public': 1}
l2_10['ctype_num'] = l2_10['ctype'].str.lower().map(mapping)
plt.plot(l2_10['name'], l2_10['ctype_num'], marker='o')
plt.title('Companies and ctype graph')
plt.xticks(rotation=45)
plt.ylabel('0 = Private, 1 = Public')
plt.show()
```

# <span style = color:green> Bar chart
***

```python
import pandas as pd
import matplotlib.pyplot as plt

df = l1.copy()

df['name'] = df['name'].astype(str)

df['years'] = df['years'].str.extract(r'(\d+)')
df['years'] = pd.to_numeric(df['years'], errors='coerce')

df = df.dropna(subset=['name', 'years'])

df = df.head(10)

plt.figure(figsize=(10,5))
plt.bar(df['name'], df['years'])

plt.xticks(rotation=45)
plt.xlabel("Company Name")
plt.ylabel("Years")
plt.title("Top 10 Companies - Age Comparison")

plt.tight_layout()
plt.show()
```

# <span style = color:green> Pie chart
***

```python
import pandas as pd
import matplotlib.pyplot as plt

df = l1.copy()

df['name'] = df['name'].astype(str)

df['years'] = df['years'].str.extract(r'(\d+)')
df['years'] = pd.to_numeric(df['years'], errors='coerce')

df = df.dropna(subset=['name', 'years'])

df = df.head(10)

plt.figure(figsize=(8,8))
plt.pie(df['years'], labels=df['name'], autopct='%1.1f%%')

plt.title("Top 10 Companies - Age Distribution")

plt.show()
```

# <span style = color:green> Funnel chart
***

```python
import pandas as pd
import matplotlib.pyplot as plt

df = l1.copy()

df['name'] = df['name'].astype(str)

df['years'] = df['years'].str.extract(r'(\d+)')
df['years'] = pd.to_numeric(df['years'], errors='coerce')

df = df.dropna(subset=['name', 'years'])

df = df.sort_values(by='years', ascending=False).head(10)

plt.figure(figsize=(10,6))
plt.barh(df['name'], df['years'], color='skyblue')

plt.gca().invert_yaxis()

plt.xlabel("Years")
plt.title("Top 10 Companies - Funnel Style Chart")

plt.tight_layout()
plt.show()
```

# <span style = color:green> Histogram
***

```python
import pandas as pd
import matplotlib.pyplot as plt

df = l1.copy()

df['years'] = df['years'].str.extract(r'(\d+)')
df['years'] = pd.to_numeric(df['years'], errors='coerce')

df = df.dropna(subset=['years'])

plt.figure(figsize=(8,5))
plt.hist(df['years'], bins=10, color='orange', edgecolor='black')

plt.xlabel("Years")
plt.ylabel("Number of Companies")
plt.title("Distribution of Company Ages")

plt.tight_layout()
plt.show()
```
