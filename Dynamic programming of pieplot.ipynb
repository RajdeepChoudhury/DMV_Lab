```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
n1 = np.array(list(map(int, input('Enter data 1:').split())))
n2 = np. array(["Raju", "Smita", "Ishani", "Mohan"])
plt.pie(n1, labels = n2, autopct = "%1.1f%%", startangle = 200,
       pctdistance = 1.4, labeldistance = 0.5)
plt.title('Points scored by students\n', color = "Green")
plt.show()
```
