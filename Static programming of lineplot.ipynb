```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
n1 = np.array([2,3,5,7,9])
n2 = np. array([3,6,9,6,1])
plt.plot(n1,n2,marker ='*')
plt.title("Points scored by students\n",fontsize = 20, color = 'Green')
plt.grid(color = 'grey')
plt.xlabel("Number of Points")
plt.ylabel("Number of Students")
plt.show()
```
