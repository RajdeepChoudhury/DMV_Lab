```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = ['A', 'B', 'C', 'D']
bar_values = [10, 15, 7, 12]
line_values = [8, 14, 6, 10]
fig, ax = plt.subplots(1, 2, figsize=(10, 4))
ax[0].bar(x, bar_values)
ax[0].set_title("Barplot")
ax[0].set_xlabel("Category")
ax[0].set_ylabel("Values")
ax[1].plot(x, line_values, marker='o')
ax[1].set_title("Lineplot")
ax[1].set_xlabel("Category")
ax[1].set_ylabel("Values")
plt.tight_layout()
plt.show()
```
