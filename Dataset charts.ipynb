```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
d1 = pd.read_csv("alzheimers_disease_data.csv")
```

```python
d1 = d1.iloc[:32,:10]
print(d1)
```

```python
d1.to_csv("alzheimers_disease_data.csv",index = False)
```

```python
plt.plot(d1['Age'],d1['AlcoholConsumption'],marker = 'o')
plt.title("Age Vs Alcohol consumption")
plt.xlabel("Age")
plt.ylabel("Alcohol Consumption amount")
plt.show()
```

```python
b1 = plt.bar(d1['Age'],d1['AlcoholConsumption'])
plt.bar_label(b1)
plt.title("Age Vs Alcohol consumption")
plt.xlabel("Age")
plt.ylabel("Alcohol Consumption amount")
plt.show()
```

```python
plt.pie(d1['BMI'], labels = d1['Age'].index,autopct ='%1.1f%%')
plt.title("Age Vs BMI")
plt.xlabel("Age")
plt.ylabel("Alcohol Consumption amount")
plt.show()
```
