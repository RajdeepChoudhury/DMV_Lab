```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
d1 = pd.read_excel(r"C:\Users\Asus\Downloads\alzheimers_disease_data.xlsx")
```

```python
d1.head(6)
```

# <span style = color:Skyblue> Lineplot
***

```python
plt.plot(d1['Age'],d1['AlcoholConsumption'],marker = 'o')
plt.title("Age Vs Alcohol consumption")
plt.xlabel("Age")
plt.ylabel("Alcohol Consumption amount")
plt.show()
```

# <span style = color:Skyblue> Barplot
***

```python
b1 = plt.bar(d1['Age'],d1['AlcoholConsumption'])
plt.bar_label(b1)
plt.title("Age Vs Alcohol consumption")
plt.xlabel("Age")
plt.ylabel("Alcohol Consumption amount")
plt.show()
```

# <span style = color:Skyblue> Pieplot
***

```python
plt.pie(d1['EducationLevel'], labels = d1['Gender'],autopct ='%1.1f%%')
plt.title("Age Vs BMI")
plt.xlabel("Age")
plt.ylabel("Alcohol Consumption amount")
plt.show()
```

# <span style = color:Skyblue> Histogram
***

```python
b1 = plt.hist(d1['AlcoholConsumption'])
plt.title("Alcohol consumption data")
plt.xlabel("Concentration level")
plt.ylabel("Alcohol Consumption amount")
plt.show()
```

# <span style = color:Skyblue> Stairplot
***

```python
plt.step(d1['Age'],d1['AlcoholConsumption'])
plt.title("Age Vs Alcohol consumption")
plt.xlabel("Age")
plt.ylabel("Alcohol Consumption amount")
plt.show()
```

# <span style = color:Skyblue> Scatterplot
***

```python
plt.scatter(d1['BMI'],d1['AlcoholConsumption'], label = 'BMI',color = 'green')
plt.legend()
plt.title("BMI Vs Alcohol consumption")
plt.xlabel("BMI Index")
plt.ylabel("Alcohol Consumption amount")
plt.show()
```

# <span style = color:Skyblue> Boxplot
***

```python
plt.boxplot(d1['BMI'])
plt.title("Boxplot of BMI data")
plt.ylabel("BMI")
plt.show()
```
