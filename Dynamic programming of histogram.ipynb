```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
n1 = np.array(list(map(int, input('Enter data 1:').split())))
n2 = np.array(list(map(int, input('Enter data 2:').split())))
plt.hist([n1,n2],label=['Data 1','Data 2'],bins=5,alpha=0.7)
plt.grid(color = 'grey')
plt.legend()
plt.title("Points scored by students\n",fontsize = 20, color = 'Green')
plt.xlabel("Number of Points")
plt.ylabel("Number of Students")
plt.show()
```
