```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation
x=0
y=0
dx=0
dy=0
r=0.2
fig,ax=plt.subplots()
ax.set_xlim(-5,5)
ax.set_ylim(-5,5)
ax.set_aspect("equal")
t=np.linspace(0,2*np.pi,200)
circle,=ax.plot([],[])
def draw():
    X=x+r*np.cos(t)
    Y=y+r*np.sin(t)
    circle.set_data(X,Y)
def update(i):
    global x,y
    x+=dx
    y+=dy
    draw()
    return circle,
def press(e):
    global dx,dy,r
    if e.key=="left": dx=-0.1
    if e.key=="right": dx=0.1
    if e.key=="up": dy=0.1
    if e.key=="down": dy=-0.1
    if e.key==" ": r+=0.05
def release(e):
    global dx,dy
    if e.key in ["left","right"]: dx=0
    if e.key in ["up","down"]: dy=0
fig.canvas.mpl_connect("key_press_event",press)
fig.canvas.mpl_connect("key_release_event",release)
draw()
anim=FuncAnimation(fig,update,interval=30,cache_frame_data=False)
plt.show()
```
