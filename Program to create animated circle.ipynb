# <span style = color:Skyblue> Program to create animated circle
***

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML

fig, ax = plt.subplots()

ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.set_aspect('equal')

cx, cy = 5, 5
r = 3

ball = plt.Circle((cx + r, cy), 0.3, color='blue')
ax.add_patch(ball)

def animate(frame):
    theta = np.radians(frame)
    x = cx + r * np.cos(theta)
    y = cy + r * np.sin(theta)
    ball.center = (x, y)
    return ball,

ani = animation.FuncAnimation(
    fig,
    animate,
    frames=360,
    interval=20,
    blit=False
)

plt.close(fig)

HTML(ani.to_jshtml())
```
